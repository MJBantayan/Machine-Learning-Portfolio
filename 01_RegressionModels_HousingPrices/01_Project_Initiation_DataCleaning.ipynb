###**2. Data Cleaning and Wrangling:**

---
Clean the data by dropping unnecessary columns and imputing missing values.




####**2.1. Dropping Columns with majority missing values:**

```python
housing_train = df_train.copy()

housing_train.drop(columns=['Id'], inplace=True)
housing_train.head()
```

Check which columns have missing values:

```python
missing_counts = housing_train.isna().sum()

for column, count in missing_counts.items():
  if count > 0:
    print(column, count)
```

Drop columns with 90% and above missing data since any imputation will have very unreliable results:

```python
housing_train.drop(columns = ["Alley", "MiscFeature", "Fence"], inplace=True)
housing_train.shape
```

Despite three digit missing values we will keep `LotFrontage`, `PoolQC` and `FirepalceQu` because it is still possible to impute them by extrapolating from related columns. Also, missing ≠ unimportant.

```python
missing_counts = housing_train.isna().sum()

for column, count in missing_counts.items():
  if count > 0:
    print(column, count)
```

####**2.2. Imputing Missing Values**
---
In this section of the notebook we will impute the missing values and save it to an imputed dataframe.



Impute `Electrical` with the most common value, `SBrkr` which is the mode of the column.

```python
housing_train['Electrical'].fillna(housing_train['Electrical'].mode()[0], inplace=True)
```

We will base the value we fill in for `PoolQC` on `PoolArea` because those houses with no data for pool quality are those with no swimming pool installed in the first place!

```python
housing_train['PoolQC'] = housing_train.apply(lambda x: 'NA' if x['PoolArea'] == 0 else x['PoolQC'], axis=1)
```

I checked reddit and I found out that houses on the same zoning and same class usually have the same lot frontage so I decided to group the missing by both categories and impute with class median:

```python
housing_train['LotFrontage'] = housing_train.groupby(['MSSubClass', 'MSZoning'])['LotFrontage'].transform(lambda x: x.fillna(x.median()))
```

The same with pool quality, fireplace quality is most likely left blank by people who answered 0 to number of fireplaces in their listing. So we impute `FireplaceQu` by assigning *NA* for *No Fireplace* for rows where `Fireplaces` has the value 0.

```python
housing_train['FireplaceQu'] = housing_train.apply(lambda x: 'NA' if x['Fireplaces'] == 0 else x['FireplaceQu'], axis=1)
```

Imputing `GarageType` and garage-related columns:

```python
df_garage = housing_train[['GarageType', 'GarageYrBlt', 'GarageFinish', 'GarageCars', 'GarageArea', 'GarageQual', 'GarageCond']].copy()
df_garage.head(15)
```

```python
df_garage.isna().sum()
```

```python
df_garage.isna().sum(axis=1).value_counts()
```

```python
temp = housing_train[housing_train['GarageType'].isna()][
    ['GarageType','GarageYrBlt', 'GarageFinish', 'GarageQual', 'GarageCond']
]
temp.head(81)
```

This confirms my suspicion that all the missing values in garage-related data are in the same rows, those that align with no garage built. We impute this by filling the rows in with 'NA' which is the value for no garage according to `data_description.txt`:

```python
cols = ['GarageType', 'GarageYrBlt', 'GarageFinish','GarageQual', 'GarageCond']
housing_train[cols] = housing_train[cols].fillna('NA')
```

```python
missing_counts = housing_train.isna().sum()

for column, count in missing_counts.items():
  if count > 0:
    print(column, count)
```

Impute the basement-related variables:

```python
df_basement = housing_train[['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'BsmtFinSF1', 'BsmtFinSF2','BsmtUnfSF', 'TotalBsmtSF']].copy()
df_basement.head(10)
```

```python
df_basement.isna().sum()
```

```python
temp2 = housing_train[housing_train['BsmtExposure'].isna()][
    ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'BsmtFinSF1', 'BsmtFinSF2','BsmtUnfSF', 'TotalBsmtSF']
]
temp2.head(38)
```

If the total basement area is 0 sq ft. then there's no basement! We will impute these area by setting the value to *NA* for each row:

```python
cols = ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']

housing_train[cols] = housing_train.apply(
    lambda x: ['NA']*len(cols) if x['TotalBsmtSF'] == 0 else x[cols], axis=1, result_type='expand')
```

Fill in the missing value for basement exposure in row 984 with the mode for the class:

```python
exp_mode = housing_train['BsmtExposure'].mode()[0]
housing_train['BsmtExposure'].fillna(exp_mode, inplace=True)
```

```python
cols = ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']
housing_train[cols].isna().sum()
```

```python
temp3 = housing_train[housing_train['BsmtFinType2'].isna()][
    ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'BsmtFinSF1', 'BsmtFinSF2','BsmtUnfSF', 'TotalBsmtSF']
]
temp3.head()
```

If there's a square feet value for finished basement `BsmtFinSF2` that means that `BsmtFinType2` is not NA! We impute this by filling in with the mode:

```python
bsmt2_mode = housing_train['BsmtFinType2'].mode()[0]
housing_train['BsmtFinType2'].fillna(bsmt2_mode, inplace=True)
```

Check where masonry veneer area is empty:

```python
df_masonry = housing_train[['MasVnrType', 'MasVnrArea']].copy()
temp4 = housing_train[housing_train['MasVnrArea'].isna()][
    ['MasVnrType', 'MasVnrArea']
]
temp4.head(8)
```

Where masonry veneer type is empty:

```python
temp5 = housing_train[housing_train['MasVnrType'].isna()][
    ['MasVnrType', 'MasVnrArea']
]
temp5.head()
```

Masonry veneer type is NA on rows where masonry area is 0 since that means no veneer was installed by the homeowners!

```python
housing_train.loc[housing_train['MasVnrArea'] == 0, 'MasVnrType'] = 'None'
```

```python
missing_counts = housing_train.isna().sum()

for column, count in missing_counts.items():
  if count > 0:
    print(column, count)
```

```python
mask_both_nan = housing_train['MasVnrType'].isna() & housing_train['MasVnrArea'].isna()
housing_train.loc[mask_both_nan, 'MasVnrType'] = 'None'
housing_train.loc[mask_both_nan, 'MasVnrArea'] = 0.0
```

```python
masvnr_mode = housing_train['MasVnrType'].mode()[0]
mask_type_nan = housing_train['MasVnrType'].isna() & housing_train['MasVnrArea'].notna()
housing_train.loc[mask_type_nan, 'MasVnrType'] = masvnr_mode
```

```python
total_missing = housing_train.isna().sum().sum()
print(total_missing)
```
